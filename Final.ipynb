```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/world_development_data_imputed.csv').dropna()
df['Year'] = df['Year'].astype(int)
df['GDPperCapita'] = df['GDP'] / df['PopTotal']
df[df['Country'] == 'Kuwait']
desired_order = ['Year', 'GDP', 'GDPperCapita', 'Country', 'Region', 'SubRegion', 'SurfAreaSqKm', 'PopTotal',
                 'PopDens', 'PopGrowth%', 'GDPGrowth%', 'AdolFertRate', 'AgriValAdd%GDP', 'Exports%GDP',
                 'FertRate', 'FDINetBoP', 'GNI/CapAtlas', 'GNIAtlas', 'Imports%GDP', 'IndValAdd%GDP',
                 'InflConsPric%', 'LifeExpBirth', 'MerchTrade%GDP', 'MobileSubs/100', 'MortRateU5',
                 'NetMigr', 'UrbanPopGrowth%']

df = df.reindex(columns=desired_order)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df2021 = df[df['Year'] == 2021]
correlation_matrix = df2021.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Variables')
plt.show()
```
